# Python Logical Operators
##### Logical operators are used to combine conditional statements. Python has three logical operators:

##### 1) **and**- Returns True if both statements are true
##### 2) **or** - Returns True if one of the statements is true
##### 3) **not** - Reverses the result, returns False if the result is true

# 1) **The  and Operator**
##### This is used to combine multiple conditions. Both conditions must evaluate to true for the expression to be true

```python
# Example 1  --> Test if A is greater than B and if C is greater than A:

a = 200
b = 33
c = 500

if a > b and c > a:
    print("True")
else:
    print("False")
```

```python
# Example 2 --> Test if wife name is Samira and son name is Haider
wife_name = input("Enter your wifes name: ")
son_name = input("Enter your son's name: ")

if wife_name == "Samira" and son_name == "Haider":
    print("True this is my family")
else:
    print("Not sure who this people are")
```

# **The or Operator**
##### This is used to combine conditional statements. At least one condition must evaluate to true for the entire expression to be true

```python
# Example 1 --> Test if A is greater than b, OR if A is greater than C
print(f"A is {a}, B is {b} and C is {c}")
if a > b or a > c:
    print("True")
else:
    print("False")
```

```python
# Example 2 --> Test if user is logged in or user session expired. 
user_logged_in = True
session_expired = False

if user_logged_in or session_expired:
    print("User Active")
else:
    print("Session expired!")
```

# **The not Operator**
##### This is also a logical operator. Its used to reverse the result of the conditional statement. The **not** keyword come before the the evaluated condition

```python
# Example 1 --> Check if a value if false 
value =  True

if not value:
    print("Value is False")
else:
    print("Value is True")
```

```python
# Example 2 --> Test if A is not greater than B:
a = 33
b = 200

if not a > b:
    print("A is NOT greater than B")
else:
    print("A is greater than B")
```

# **Combining Multiple Operators**
##### You can combine multiple logical operators in a single expression. Python evaluates not first, then and, then or. 
##### When combining multiple logical operators, use parentheses to make your intentions clear and control the order of evaluation.



```python
# Example 1
age = 25
is_student = False
has_discount_code = True

if (age < 18 or age > 65) and not (is_student or has_discount_code):
  print("Discount applies!")
else:
  print("No discount available.")
```

```python
# Example 2 --> Use authentication check
username = "Tobias"
password = "secret123"
is_verified = True

if username and password and is_verified:
    print("Login successful!")
else:
    print("Login failed. Please check your credentials and verification status.")
```

```python
# Example 3 --> Range checking with logical operators:
score = 85

if score >= 0 and score <= 100:
    print("Score is valid.")
else:
    print("Score is invalid.")
```

```python
# Example 4 --> Use an if statement to print "YES" if either a or b is equal to c.
a = 2
b = 50
c = 2

if a == c or b == c:
    print("YES")
```
