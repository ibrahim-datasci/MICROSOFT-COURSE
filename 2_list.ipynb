# List
### This chapter covers 
#### 1) List Creation
#### 2) Adding elements to a list
#### 3) Removing elements from a list
#### 4) Indexing
#### 5) List Unpacking

```python
my_skills  = ["Python", "SQL", "Machine Learning", "Data Visualization"]
```

```python
print(my_skills)
```

```python
# Appending a new skill to the list
my_skills.append("Deep Learning")
```

```python
print(my_skills)
```

```python
# Getting the length of the list
print(len(my_skills))
```

```python
# Indexing list elements
print(my_skills[0]) # This will print the first element of the list
```

```python
# Slicing the list
print(my_skills[1:3]) # This will print the second and the third elements of the list. The last index is exclusive.
```

```python
# Adding steps in the slicing
print(my_skills[0:4:2]) # This will print the first and the third elements of the list. The last index is exclusive and the step is 2, which means it will skip every second element.
```

```python
# Removing a skill from the list
my_skills.remove("SQL")
```

```python
print(my_skills)
```

```python
# Removing and returning the last element of the list
last_skill = my_skills.pop()
```

```python
print(last_skill)
```

```python
# Removing a skill from the list using the pop method with an index
second_skill = my_skills.pop(1)
```

```python
second_skill
```

```python
print(my_skills)
```

```python
# Creating a new list of school names
schools = ["Harvard", "MIT", "Stanford", "Yale"]
```

```python
print(schools)
```

```python
my_school_choice, *other_schools = schools
```

```python
print(my_school_choice)
print(other_schools)
```

```python
# Method 2 of skillfully unpacking a list
school_1, schools = schools[0], schools[1:]
```

```python
print(school_1)
print(schools)
```
