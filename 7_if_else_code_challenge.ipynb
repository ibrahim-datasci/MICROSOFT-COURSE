# Challenge 5
### Write a program that:

### Asks the user for:
#### age
#### score

## Handles errors if the user enters something that is not a number

#### Then applies this logic:
#### 📋 Rules
##### If age is less than 18 → print: "Too young"

#### If age is 18 or more, then check score:
##### score ≥ 80 → "Eligible - Excellent"
##### score ≥ 50 → "Eligible - Pass"
##### score < 50 → "Eligible - Fail"

```python
try:
    user_age = int(input("Enter your age: "))
    user_score = int(input("Enter your score: "))
    if user_age < 18:
        print("Too young")
    else:
        if user_score >= 80:
            print("Eligible - Excellent")
        elif user_score >= 50:
            print("Eligible - Pass")
        else: 
            print("Eligible - Fail")
except ValueError:
    print("One or both values provided not a number")
```
