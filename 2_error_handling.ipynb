# Error Handling and Debugging
##### Error handling - is when you have errors that are not in your control
##### Debugging is when you know there is a problem with your code.
##### Syntax Error - Our code is not going to run at all. This is the easiest error to try and track down. The error messages will typically point you to where the error is. 
##### Runtime Errors - Code will run, something goes wrong and it will blow out. This will also point at exactly where the problem is. 
##### Logic Error - Code compiles perfectly. No sytax error. No error message is going to be printed. Everything runs. 

##### Try and Except is not used to find bugs. Also you dont have to catch all error. 

```python
# Example of a syntax error. This code wont run
x = 42
y = 206

if x == y
    print("Success!")
```

```python
# Runtime error
x = 42
y = 0
print(x/y)
```

```python
# Catching Runtime Errors
x = 40
y = 0
try:
    print(x/y)
except ZeroDivisionError as e:
    print("Error: " + str(e))``
else:
    print("Something else went wrong")
finally:
    print("This code will always run")
```

```python
# Logic Error. This is the most common error people make when they are learning to code. 
# The code runs without any errors, but the output is not what you expected.
if x < y:
    print(str(x) + " is greater than " + str(y))
```
