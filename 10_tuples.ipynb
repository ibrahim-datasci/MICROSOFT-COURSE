# Tuples
`Tuples` are used to store multiple items in a single variable.

`Tuple` is one of 4 built-in data types in Python used to store collections of data, the other 3 are `List`, `Set`, and `Dictionary`, all with different qualities and usage.

A `tuple` is a collection which is **ordered** and **unchangeable**.

`Tuples` are written with round brackets.

### Ordered
When we say that tuples are ordered, it means that the items have a defined order, and that order will not change.

### Unchangeable
Tuples are unchangeable, meaning that we cannot change, add or remove items after the tuple has been created.

### Allow Duplicates
Since tuples are indexed, they can have items with the same value:

```python
# Creating a tuple
myfirst_tuple = ("Apple", "TV", "Haider", "Apple")
myfirst_tuple
```

```python
myfirst_tuple[0]
```

### Create Tuple With One Item
To create a tuple with only one item, you have to add a `comma` after the item, otherwise Python will not recognize it as a tuple.

```python
thistuple = ("Apple",)
print(type(thistuple))
```

```python
# Not a tuple if you dont add a comma at the end
not_tuple = ("Apple")
print(type(not_tuple))
```

```python
x = ("apple", "banana", "cherry")
x[1] = "Water mellon"
```

### Unpack Tuples
When we create a tuple, we normally assign values to it. This is called "packing" a tuple:

```python
fruits = ("Kiwi", "Banana", "Cherry")
fruits
```

```python
fruit1, fruit2, fruit3 = fruits
print(fruit1)
print(fruit2)
print(fruit3)
```

* **Note: The number of variables must match the number of values in the tuple, if not, you must use an asterisk to collect the remaining values as a list.**

### Using Asterisk*
* If the number of variables is less than the number of values, you can add an * to the variable name and the values will be assigned to the variable as a list:
* If the asterisk is added to another variable name than the last, Python will assign values to the variable until the number of values left matches the number of variables left

```python
fruits = ("apple", "banana", "cherry", "strawberry", "raspberry")
fruits
```

```python
green,yellow, *red = fruits
print(green)
print(yellow)
print(red)
```

```python
# Example 2 --> * other than the last variable
fruits

green, *tropic, red = fruits
print(green)
print(tropic)
print(red)
```

### Looping through a tuple
You can loop through the tuple items by using a for loop.


```python
thistuple = ("apple", "banana", "cherry")
for fruit in thistuple:
    print(fruit)
```

```python
# Example 2 --> Looping through the Index Numbers
thistuple = ("apple", "banana", "cherry")

for x in range(len(thistuple)):
    print(thistuple[x])
```

```python
## Uisng  a while loop
thistuple = ("apple", "banana", "cherry")
i = 0

while i < len(thistuple):
    print(thistuple[i])
    i+=1
```

### Join Two Tuples
To join two or more tuples you can use the + operator:

```python
# Example 1 -> Using the + Operator

tupleone = ("Apple", "Banana", "Kiwi")
tupletwo = ("Orange", "Mangoe")
tuplethree = tupleone + tupletwo
tuplethree
```

#### Multiply Tuples
If you want to multiply the content of a tuple a given number of times, you can use the * operator:

```python
# Example 1
fruits = ("apple", "banana", "cherry")
mytuple = fruits * 2

mytuple
```

```python
# Example 2
tuple1 = (1,2,3,4,5)
tuple2 = tuple1 * 2

tuple2
```
