# **Conditional Logic**
### Python supports the usual logical conditions from mathematics:

#### Equals: a == b
#### Not Equals: a != b
#### Less than: a < b
#### Less than or equal to: a <= b
#### Greater than: a > b
#### Greater than or equal to: a >= b

### These conditions can be used in several ways, most commonly in "if statements" and loops.

### An "if statement" is written by using the if keyword.

## **Indentation**
#### Python relies on indentation (whitespace at the beginning of a line) to define scope in the code. Other programming languages often use curly-brackets for this purpose.

```python
name = "Samira"
```

```python
# First if statement
if name == "Samira":
    print("Hello Samira!")
```

```python
# This will throw an indentation error. After the if statement, you need to indent the code that should run if the condition is true.
a = 33
b = 200
if b > a:
print("b is greater than a")
```

```python
# Multiple Statements in If Block. Both print statements will only run if the condition is true.
new_age = 30
if new_age >= 18:
    print("You are an adult.")
    print("You can vote.")
```

# The Elif Keyword
##### The elif keyword is Python's way of saying "if the previous conditions were not true, then try this condition".
##### The elif keyword allows you to check multiple expressions for True and execute a block of code as soon as one of the conditions evaluates to True.

```python
if new_age >= 18:
    print("You are an adult.")
elif new_age < 18 and new_age >= 13:
    print("You are a teenager.")
```

# T**he Else Keyword**
##### The else keyword catches anything which isn't caught by the preceding conditions.
##### The else statement is executed when the if condition (and any elif conditions) evaluate to False.
##### The else statement acts as a fallback that executes when none of the preceding conditions are true. This makes it useful for error handling, validation, and providing default values.

```python
try:
    new_age = int(input("Please enter your age: "))
except ValueError as e:
    print("Invalid input: " + str(e))
finally:
    print("This code will always run")


if new_age >= 18:
    print("You are an adult.") 
elif new_age < 18 and new_age >= 13:
    print("You are a teenager.")
else:
     print("You are a child.")
```

# Python Shorthand If
##### If you have only one statement to execute, you can put it on the same line as the if statement.
##### **Note: You still need the colon : after the condition.**



```python
a = 5
b = 2

if a > b: print("a is greater than b")
```

# Short Hand If ... Else
##### If you have one statement for if and one for else, you can put them on the same line using a conditional expression:


```python
# Example 1
wife_name = "Samira"
print("Hello " + wife_name + "!") if wife_name == "Samira" else print("Hello stranger!")
```

```python
# Example 2
x = 10
y = 5

print(f"{x} is greater than {y}") if x > y else print(f"{x} is less than or equal to {y}")
```

# Assign a Value With If ... Else
##### You can also use a one-line if/else to choose a value and assign it to a variable:

```python
# Example 1
a = 40
b = 80

greater_value = a if a > b else b
print(f"The greater value is: {greater_value}")
```

```python
# Example 2
wife_name = "Samira"
other_lady = "Jane"

true_name_wife = wife_name if wife_name == "Samira" else other_lady 
print(f"Hello {true_name_wife}!")
```

# Multiple Conditions on One Line
##### You can chain conditional expressions, but keep it short so it stays readable:
##### NOTE: We dont use elif statement

```python
a = 300
b = 600

print("A") if a > b else print("B") if b > a else print("A and B are equal")
```

# PRACTICAL EXAMPLES
##### Ternary operators are particularly useful for simple assignments and return statements.

```python
# Example 1 - User welcome custome message
user_name = input(str("Please enter your name: "))
display_name = user_name if user_name else "Guest"
print(f"Hello {display_name}!")
```

```python
# Example 2 - Finding the maximum of two numbers:
num1 = int(input("Enter the first number: "))
num2 = int(input("Enter the second number:"))
max_num = num1 if num1 > num2 else num2
print(f"The maximum number is: {max_num}")
```

# When to Use Shorthand If
### Shorthand if statements and ternary operators should be used when:

##### a) The condition and actions are simple
##### b) It improves code readability
##### c) You want to make a quick assignment based on a condition

##### ***Important: While shorthand if statements can make code more concise, avoid overusing them for complex conditions. For readability, use regular if-else statements when dealing with multiple lines of code or complex logic.***
