# The Python Match Statement

##### Instead of writing many if..else statements, you can use the match statement.

##### The match statement selects one of many code blocks to be executed

### **SYNTAX**
##### match expression:
  ##### case x:
     code block
  ##### case y:
     code block
  ##### case z:
    code block


```python
day = 4
match day:
    case 1:
        print("Monday")
    case 2: 
        print("Tuesday")
    case 3:
        print("Wednesday")
    case 4:
        print("Thursday")
    case 5:
        print("Friday")
    case 6:
        print("Saturday")
    case 7:
        print("Sunday")
```

## Default Value
##### Use the underscore character _ as the last case value if you want a code block to execute when there are not other matches:This is basically like the final else statement

```python
day = 5
match day:
    case 1:
        print("Monday")
    case 2:
        print("Tuesday")
    case 3:
        print("Wednesday")
    case 4:
        print("Thurday")
    case 5: 
        print("Friday")
    case _:
        print("Weekend")
```

### Comine Values
##### Use the | character as an OR operator in the case evaluation to check for more than one value match in one case:

```python
day = 6
match day:
    case 1 | 2 | 3 | 4 | 5:
        print("Weekday")
    case _:
        print("Weekend")
    
```

### if statement as Guards:
##### You can add if statements in the case evaluation as an extra condition-check:

```python
day = 5
month = 4

match day:
    case 1 | 2 | 3 | 4 | 5 if month == 4:
        print("Weekday in April")
    case 1 | 2 | 3 | 4 | 5 if month == 5:
        print("Weekday in May")
    case _:
        print("No match")
```

### Quiz 1: Day of the Week

Write a program that:

- Asks the user to enter a number (1–7)
- Uses a `match` statement to print the corresponding day:

1 → Monday  
2 → Tuesday  
3 → Wednesday  
4 → Thursday  
5 → Friday  
6 → Saturday  
7 → Sunday  

If the number is outside 1–7, print:
"Invalid day"

```python
try: 
    number = int(input("Enter a number betweeen 1 and 7: "))
    match number:
        case 1:
            print("Monday")
        case 2: 
            print("Tuesday")
        case 3:
            print("Wednesday")
        case 4:
            print("Thursday")
        case 5: 
            print("Friday")
        case 6:
            print("Saturday")
        case 7: 
            print("Sunday")
        case _:
            print("Invalid number!")
except ValueError:
    print("Value provided not a number")
```

### Quiz 2: Simple Calculator

Write a program that:

- Asks the user for:
  - First number
  - Second number
  - An operation (+, -, *, /)

- Uses a `match` statement to perform the operation

Rules:
- "+" → addition  
- "-" → subtraction  
- "*" → multiplication  
- "/" → division  

If the operation is not one of the above, print:
"Invalid operation"

```python
valid_operations = ['+', '-', '*', '/']
try:
    first_number = int(input("Enter first number: "))
    second_number = int(input("Enter second number: "))
    operation = input("Enter an operation")
    match operation:
        case "+":
            print(first_number + second_number)
        case "-":
            print(first_number - second_number)
        case "*":
            print(first_number * second_number)
        case "/":
            if second_number != 0:
                print(first_number / second_number)
            else: 
                print("cannod divded first number by 0")
        case _:
            print("Not a valid operation")
except ValueError:
    print("Value provided not a number")
```

### Quiz 3: Grade Feedback System

Write a program that:

- Asks the user to enter a grade letter:
  (A, B, C, D, F)

- Uses a `match` statement to print feedback:

A → "Excellent"  
B → "Good"  
C → "Average"  
D → "Below Average"  
F → "Fail"  

If the grade is anything else, print:
"Invalid grade"

```python
grade = input("Enter a grade: ").upper()
match grade:
    case "A":
        print("Excellent")
    case "B":
        print("Good")
    case "C":
        print("Average")
    case "D":
        print("Below Average")
    case "F":
        print("Fail")
    case _:
        print("Invalid grade")
```
