# Python While Loops
With the while loop we can execute a set of statements as long as a condition is true.

```python
# Example 1
#Note: remember to increment i, or else the loop will continue forever.
i = 1

while i < 6:
    print(i)
    i+=1


```

# The break Statement
With the `break` statement we can stop the loop even if the while condition is true:

```python
i = 0

while i < 5:
    print(i)
    if i == 3:
        print("We are breaking here")
        break
    i+=1
```

# The continue Statement
 With the `continue` statement we can stop the current iteration, and continue with the next:

```python
i = 0
while i < 5:
    i+=1
    if i == 3:
        continue
    print(i)
```

# The else Statement
With the `else` statement we can run a block of code once when the condition no longer is true: 
**Note:** The else block will NOT be executed if the loop is stopped by a break statement.



```python
i = 1

while i < 6:
    print(i)
    i+=1
else:
    print("i is no longer less than 6")
```

### Quiz 1: Countdown Timer

Write a program that:

- Asks the user to enter a number
- Uses a `while` loop to count down from that number to 1

Rules:
- Print each number on a new line
- When the countdown finishes, print:
  "Done!"

Example:

Input:
5

Output:
5
4
3
2
1
Done!

```python
try:
    number = int(input("Enter a number"))
    while number >= 1:
        print(number)
        number-=1
    print("Done")
except ValueError as e:
    print(f"Error: {e}")
```

### Quiz: Login System With Attempt Limit

Write a program that:

- Has a predefined password: `"python123"`
- Allows the user **only 3 attempts** to enter the correct password
- Uses a `while` loop

### Rules:

1. The user gets a maximum of **3 attempts**
2. If the password is correct:
   - Print `"Access granted"`
   - Stop the program immediately
3. If the password is wrong:

   - Reduce remaining attempts by 1
4. If all 3 attempts are used:

### Another Example:

Input:
abc  
123  
xyz  

Output:
Incorrect password  
Incorrect password  
Incorrect password  
Account locked  

```python
valid_password = "python123"
attempts = 0

while True:
    user_input = input("Enter your password: ")
    if user_input == valid_password:
        print("Access granted")
        break
    else:
        print("Incorrect password")
        attempts += 1
        if attempts >= 3:
            print("Account locked")
            break


```
